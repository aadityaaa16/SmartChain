```python
import requests
from transformers import pipeline
import pandas as pd

NEWS_API_KEY = "9f7bc011f71d4bf09c26c80330b46e37"

# Load FinBERT
print("Loading FinBERT model")
sentiment_analyzer = pipeline("sentiment-analysis", model="ProsusAI/finbert")
print("Model loaded!")

companies = ["Tata Steel", "Maersk", "Foxconn", "Tesla", "Samsung"]

results = []

for company in companies:
    print(f"Fetching news for {company}...")
    url = f"https://newsapi.org/v2/everything?q={company}&language=en&sortBy=publishedAt&pageSize=3&apiKey={NEWS_API_KEY}"
    response = requests.get(url)
    data = response.json()

    if data.get("status") != "ok":
        print(f"Error for {company}: {data.get('message')}")
        continue

    articles = data.get("articles", [])

    for article in articles:
        headline = article["title"]
        if not headline:
            continue
        sentiment = sentiment_analyzer(headline)[0]
        results.append({
            "company": company,
            "headline": headline,
            "sentiment": sentiment["label"],
            "confidence": round(sentiment["score"], 3),
            "published": article["publishedAt"]
        })

df = pd.DataFrame(results)
print("\nResults:")
print(df)

df.to_csv("real_supplier_sentiment.csv", index=False)
print("\nSaved to real_supplier_sentiment.csv")
```
